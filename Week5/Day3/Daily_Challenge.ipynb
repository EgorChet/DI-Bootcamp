#### 1. Import the numpy, pandas, and matplotlib.pyplot packages.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### 2. Use numpy to create an array x of 100 numbers ranging from 0 to 20 (inclusive) using the linspace function.

```python
x = np.linspace(0, 20, 100)
x
```


#### 3. Create an array y which is the square of x.

```python
y = x**2
y
```


#### 4. Use matplotlib to plot x vs y.


```python
plt.figure(figsize=(8, 6))
plt.plot(x, y)
```


#### 5. Your plot should have an appropriate title and labels for the x and y axes. You should also adjust the x and y limits to correctly show all the data.


```python
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x vs y (y=x^2')
plt.xlim(0, 20)
plt.ylim(0, 400)
```


#### 6. Save this plot as a .png file using savefig.

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x vs y (y=x^2')
plt.savefig('PlotOfXvsY.png') 
```

# Task 2:
#### 1. Import the dataset from this link using pandas and assign it to a variable ‘chipo’.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, delimiter='\t')
chipo.head()
```

#### 2. Clean the item_price column and transform it in a float. 
##### Remove the dollar sign ($) from the ‘item_price’ column using the apply function in combination with a lambda function. 
##### Then convert the ‘item_price’ column to the float datatype.

```python
# chipo['item_price'] = chipo['item_price'].astype(str)
# chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
# chipo['item_price'].head()

# 1. Remove the dollar sign ($) from 'item_price'
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# 2. Convert 'item_price' to float data type
chipo['item_price'] = chipo['item_price'].astype(float)

# Display the first few rows to verify the changes
chipo.head()
```

#### 3. Create a scatter plot presenting the relationship between the price of items and the quantity ordered. Use the ‘item_price’ column for the x-axis and the ‘quantity’ column for the y-axis.

```python
price = chipo['item_price']
quantity = chipo['quantity']

plt.figure(figsize=(12, 4))
plt.scatter(price, quantity)
```

#### 4. Your plot should have an appropriate title and labels for the x and y axes.

```python
price = chipo['item_price']
quantity = chipo['quantity']

plt.figure(figsize=(12, 4))
plt.scatter(price, quantity)
plt.xlabel('Item Price')
plt.ylabel('Quantity Ordered')
plt.title('Relationship between Item Price and Quantity Ordered')
```

#### 5. Save this plot as a .png file using savefig.

```python
price = chipo['item_price']
quantity = chipo['quantity']

plt.figure(figsize=(12, 4))
plt.scatter(price, quantity)
plt.xlabel('Item Price')
plt.ylabel('Quantity Ordered')
plt.title('Relationship between Item Price and Quantity Ordered')
plt.savefig('Relationship between Item Price and Quantity Ordered.png')
```

# Task 3:
#### 1. Group the ‘chipo’ data by item_name and calculate the total quantity for each item.

```python
total_quantity_for_each_item = chipo.groupby('item_name')['quantity'].sum()
total_quantity_for_each_item
```

#### 2. Create a bar plot that presents the total quantities for the top 5 items. Use the ‘item_name’ column for the x-axis and the calculated total quantity for each item for the y-axis.
#### 3. Your plot should have an appropriate title and labels for the x and y axes.
#### 4. Save this plot as a .png file using savefig.

```python
top_5_items = total_quantity_for_each_item.sort_values(ascending=False).head(5)

plt.figure(figsize=(10, 6))
top_5_items.plot(kind='bar', rot=45)
plt.title('Top 5 Items by Quantity')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity')
plt.savefig('Top 5 Items by Quantity.png')
```
