# 🌟 Exercise 7 :
### Create a line plot and scatter plot representing your favorite mathematical functions. Change the line color to tab:blue for the line plot and use a marker of your choice for the scatter plot.

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# Line Plot (sin(x) in blue)
plt.figure(figsize=(8, 4))  # Set the figure size (optional)
plt.plot(x, y1, color='tab:blue', label='sin(x)')  # Line plot with blue color
plt.title('y = sin(x)')  # Title for the plot
plt.legend()  # Display legend

# Save the line plot as an image in memory
buffer1 = BytesIO()
plt.savefig(buffer1, format='png')
buffer1.seek(0)

# Scatter Plot (cos(x) with markers)
plt.figure(figsize=(8, 4))  # Set the figure size (optional)
plt.scatter(x, y2, marker='o', color='tab:blue', label='cos(x)')  # Scatter plot with markers
plt.title('y = cos(x)')  # Title for the plot
plt.legend()  # Display legend

# Save the scatter plot as an image in memory
buffer2 = BytesIO()
plt.savefig(buffer2, format='png')
buffer2.seek(0)


plt.show()  # Display the plots (optional)
```

# 🌟 Exercise 8 & 9 
### Plot the line plot and scatter plot from Exercise 7 on the same figure using subplots. Use plt.subplots to create a figure with two subplots.
### Modify the plots from Exercise 8 to include a legend. Use the legend function from Matplotlib.

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Display the line plot from buffer1 on the first subplot
ax1.imshow(plt.imread(BytesIO(buffer1.read())), aspect='auto')
ax1.axis('off')  # Turn off axis labels and ticks (optional)

# Display the scatter plot from buffer2 on the second subplot
ax2.imshow(plt.imread(BytesIO(buffer2.read())), aspect='auto')
ax2.axis('off')  # Turn off axis labels and ticks (optional)

plt.tight_layout()  # Adjust spacing between subplots
plt.show()  # Display the subplots
```
